```python
import numpy as np
import matplotlib.pyplot as plt

def generate_product_prices(num_prices=100, mean=7.5, std_dev=0.75, min_price=5, max_price=10):
    # Generate normally distributed prices
    prices = np.random.normal(loc=mean, scale=std_dev, size=num_prices)

    # Clip the prices to ensure they are within the range [min_price, max_price]
    prices = np.clip(prices, min_price, max_price)

    # Round the prices to 2 decimal places
    prices = np.round(prices, 2)

    return prices

# Example usage
num_prices = 100  # Number of product prices to generate
mean = 7.5        # Mean of the normal distribution (middle of the 5 to 10 range)
std_dev = 0.2     # Standard deviation (spread of prices)
min_price = 5      # Minimum price
max_price = 10     # Maximum price

# Generate the product prices
prices = generate_product_prices(num_prices, mean, std_dev, min_price, max_price)

# Print the first 10 prices as a sample
print("Sample of generated product prices:", prices[:10])

# Plot the histogram of the generated prices
plt.hist(prices, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Generated Product Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
