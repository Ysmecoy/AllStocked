```python
import numpy as np
import matplotlib.pyplot as plt

def generate_bimodal_scores(num_scores=100, mean1=40, mean2=75, std_dev1=5, std_dev2=10, ratio1=0.5):
    # Generate scores for the first mode (mean1)
    scores1 = np.random.normal(loc=mean1, scale=std_dev1, size=int(num_scores * ratio1))

    # Generate scores for the second mode (mean2)
    scores2 = np.random.normal(loc=mean2, scale=std_dev2, size=num_scores - len(scores1))

    # Combine both distributions
    scores = np.concatenate([scores1, scores2])

    # Clip the scores to ensure they are within the range [0, 100]
    scores = np.clip(scores, 0, 100)

    # Round to integers to simulate test scores
    scores = np.round(scores).astype(int)

    return scores

# Example usage
num_scores = 100  # Number of test scores to generate
mean1 = 40         # Mean of the first mode
mean2 = 75         # Mean of the second mode
std_dev1 = 5       # Standard deviation of the first mode
std_dev2 = 5      # Standard deviation of the second mode
ratio1 = 0.5       # Proportion of scores from the first mode

# Generate bimodal test scores
scores = generate_bimodal_scores(num_scores, mean1, mean2, std_dev1, std_dev2, ratio1)

# Print the first 10 scores as a sample
print("Sample of generated scores:", scores[:10])

# Plot the histogram of the generated scores
plt.hist(scores, bins=range(0, 101, 1), edgecolor='black', alpha=0.7)
plt.title('Bimodal Distribution of Generated Test Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.xticks(range(0, 101, 10))
plt.grid(True)
plt.show()
```
